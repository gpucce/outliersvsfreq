```python
import matplotlib.pyplot as plt
import torch
import sys
sys.path.append("..")
from outlier_analysis.pyplot.plotting_correlation import stack_plot_corr, stack_plot_freq
```

```python
freqs = torch.load("../outlier_correlation/output/freqs.bin")
hidden_states_special = torch.load("../outlier_correlation/output/hidden_states_special.bin")
attention_special = torch.load("../outlier_correlation/output/attentions_special.bin")
masks_special = torch.load("../outlier_correlation/output/masks_special.bin")
```

```python
stack_plot_freq(
    freqs, {j:[torch.abs(i) for i in k] for j,k in hidden_states_special.items()}, masks_special,
    outliers_idxs=[308, 381], random_dims=set(hidden_states_special.keys()).difference([308, 381])
)
```

```python
stack_plot_corr(
    {j:[torch.abs(i) for i in k] for j,k in hidden_states_special.items()}, attention_special, masks_special, 
    outliers_idxs=[308, 381], random_dims=set(hidden_states_special.keys()).difference([308, 381])
)
```
